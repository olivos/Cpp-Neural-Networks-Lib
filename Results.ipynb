```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Using 590 assets, stochastic gradient descent of 10 epochs, varying the batchsize

```python
batchStats = pd.read_csv("batchStats.csv")
batchStats
fig1,ax1 = plt.subplots()
ax1.plot(range(1,591),batchStats['time'])
ax1.set_title('Time')
ax1.set_xlabel('Batch Size')


batchStats['final gradient norm'] = np.minimum(batchStats['final gradient norm'],1e4)
fig2,ax2 = plt.subplots()
ax2.plot(range(1,591),batchStats['final gradient norm'])
ax2.set_title('final gradient norm')
ax2.set_xlabel('Batch Size')


batchStats['final error'] = np.minimum(batchStats['final error'],1e3)
fig3,ax3 = plt.subplots()
ax3.plot(range(1,591),batchStats['final error'])
ax3.set_title('final error')
ax3.set_xlabel('Batch Size')

```

```python
np.argmin(batchStats['final gradient norm']),np.argmin(batchStats['final error'])
```
